```python
import pandas as pd
import yfinance as yf
```

```python
companies = pd.read_csv('list-companies.csv')
```

```python
df = yf.Ticker(companies['ticker'][0]).history(start="2018-01-01", end="2023-12-31")
df['Ticker'] = companies['ticker'][0]
df['CIK'] = companies['CIK'][0]
for i in range(1, len(companies)):
    df2 = yf.Ticker(companies['ticker'][i]).history(start="2018-01-01", end="2023-12-31")
    df2['Ticker'] = companies['ticker'][i]
    df2['CIK'] = companies['CIK'][i]
    df = pd.concat([df, df2])
```

```python
df
```

```python


for year in df.index.year.unique():
    a = df[df.index.year == year]
    a.to_csv(f'data/{year}.csv', index=False)
```
